```python
# Imports and Helper Functions
# data Analysis
import pandas as pd
import numpy as np
import random as rng
import os

#SciKit Learn Models and libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
os.chdir('/home/vivekr/UFC_Project/data')
UFC_data = pd.read_csv("data.csv")
UFC_data.info()
```

```python
UFC_data.describe()
```

```python
UFC_data.head()
```

```python
UFC_data.fillna(value=0,inplace=True)
```

```python
UFC_data.tail()
```

```python
dropdata = UFC_data.drop(['B_ID','B_Name','R_ID','R_Name','winby','Date'],axis=1)
dropdata.rename(columns={'BPrev':'B__Prev',
                         'RPrev':'R__Prev',
                         'B_Age':'B__Age',
                         'B_Height':'B__Height',
                         'B_Weight':'B__Weight',
                         'R_Age':'R__Age',
                         'R_Height':'R__Height',
                         'R_Weight':'R__Weight',
                         'BStreak':'B__Streak',
                         'RStreak': 'R__Streak'},inplace=True)
dropdata.describe()
```

```python
dropdata.describe(include=['O'])
```

```python
objecttypes = list(dropdata.select_dtypes(include=['O']).columns)
for col in objecttypes:
    dropdata[col] = dropdata[col].astype('category')
```

```python
cat_columns = dropdata.select_dtypes(['category']).columns
dropdata[cat_columns] = dropdata[cat_columns].apply(lambda x: x.cat.codes)
dropdata.info()
dropdata.tail()
```

```python
# Subset Correlation Matrix
k = 10 #number of variables for heatmap
corrmat = dropdata.corr()
cols = corrmat.nlargest(k, 'winner')['winner'].index
cm = np.corrcoef(dropdata[cols].values.T)  #Return Pearson product-moment correlation coefficients.
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Models = dict()
X_all = dropdata.drop(['winner'], axis=1)
y_all = dropdata['winner']
X_train, X_test, Y_train, Y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=0)
randomfor = RandomForestClassifier(max_features="sqrt", n_estimators = 700, max_depth = None, n_jobs=-1)
randomfor.fit(X_train,Y_train)
Y_pred = randomfor.predict(X_test)
Models['Random Forests'] = [accuracy_score(Y_test,Y_pred),confusion_matrix(Y_test,Y_pred)]
Models
```

```python
randomfor.feature_importances_
```

```python
import sklearn.metrics as m
m.accuracy_score(Y_pred,Y_test)
```

```python
from sklearn.model_selection import validation_curve, learning_curve
train_scores,validation_scores = validation_curve(randomfor,X_all,y_all,"max_features",['auto','log2','sqrt',100,200,300,400],cv=10)
train_scores_mean = np.mean(train_scores,axis=1)
validation_scores_mean = np.mean(validation_scores,axis=1)
plt.ylim(0.0,1.1)
ticks = ['auto','log2','sqrt',1,2,3,4]
plt.xticks(range(0,7),ticks)
plt.xlabel("max_features")
plt.ylabel("Scores")
plt.plot(train_scores_mean,label="Training_scores")
plt.plot(validation_scores_mean,label="Validation Scores")
plt.legend(loc="best")
plt.title("Validation Curve")
plt.show()
```

```python
train_sizes = np.linspace(0.1,1.0,5)
train_sizes,train_scores,test_scores = learning_curve(randomfor,X_all,y_all,cv=10,train_sizes=train_sizes)
train_scores_mean = np.mean(train_scores,axis=1)
test_scores_mean = np.mean(test_scores,axis=1)
plt.grid()
plt.xlabel("Training_examples")
plt.ylabel("Scores")
plt.plot(train_scores_mean,label="Training_scores")
plt.plot(test_scores_mean,label="Test Scores")
plt.legend(loc="best")
plt.title("Learning Curve")
plt.show()
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# ROC for test data

y = label_binarize(Y_test, classes=[0,1,2,3])
n_classes = y.shape[1]
fpr,tpr,roc_auc = dict(),dict(),dict()
for i in range(n_classes):
    fpr[i],tpr[i],_ = roc_curve(y[:,i],Y_pred)
    roc_auc[i] = auc(fpr[i],tpr[i])
    
plt.figure()
lw = 2
plt.plot(fpr[3],tpr[3],lw=lw,label='ROC curve (area = %0.2f)' %roc_auc[2])
plt.plot([0,1],[0,1],color='navy',lw=lw,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

```python
# ROC for training Data
y = label_binarize(Y_train, classes=[0,1,2,3])
n_classes = y.shape[1]
fpr,tpr,roc_auc = dict(),dict(),dict()
y_score = randomfor.predict(X_train)
for i in range(n_classes):
    fpr[i],tpr[i],_ = roc_curve(y[:,i],y_score)
    roc_auc[i] = auc(fpr[i],tpr[i])
    
plt.figure()
lw = 2
plt.plot(fpr[3],tpr[3],lw=lw,label='ROC curve (area = %0.2f)' %roc_auc[2])
plt.plot([0,1],[0,1],color='navy',lw=lw,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
